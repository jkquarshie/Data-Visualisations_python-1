<p>Now let's do some data cleaning. We will deal with missing values in the dataset and do other things to prepare the data for further analysis.</p>

```python
#to be sure we have the data in view, let's take a peek
#at the first 5 rows using the ,head() method
df.head()
```

```python
df.dropna(subset = ['normalized-losses'], axis = 0, inplace = True)
```

```python
df.head()
```

```python
import numpy as  np
```

```python
df.replace('?', np.nan, inplace = True)
df.head()
```

```python
#to identify rows with missing values
missing = df.isnull()
```

```python
missing.head(5)
```

```python
for column in missing.columns.values.tolist():
    print(column)
    print(missing[column].value_counts())
    print('')
```

Based on the summary above, each column has 205 rows of data, seven columns containing missing data:
<ol>
    <li>"normalized-losses": 41 missing data</li>
    <li>"bore": 4 missing data</li>
    <li>"stroke" : 4 missing data</li>
    <li>"horsepower": 2 missing data</li>
    <li>"peak-rpm": 2 missing data</li>
    <li>"num-of-doors": 2 missing data</li>
    <li>"price": 4 missing data</li>
</ol>

```python
# now lets deal with the missing values
```

```python
# 1.  for 'normalized-losses', we have 41 missing to replace
# We replace with the mean using the astype method

#this code calculates the mean of normalized-losses column
# and converts the values to float type
avg_nl = df['normalized-losses'].astype('float').mean(axis=0)

#let's print the result to screen to view
print('Average of normalized-losses column is:', avg_nl)

#now let's use avg_nl to replace missing values in column
df['normalized-losses'].replace(np.nan, avg_nl, inplace = True)
```

```python
#now lets move on to the next missing value -'bore'

#first, we calculate and print the mean value to screen
avg_bore = df['bore'].astype('float').mean(axis =0)
print('Average of bore  column is:', avg_bore)

#next, we replace missing values with calculated mean
df['bore'].replace(np.nan, avg_bore, inplace = True)
```

```python
#next, lets move on to the next missing value -'stroke'

#first, we calculate and print the mean value to screen
avg_stroke = df['stroke'].astype('float').mean(axis =0)
print('Average of stroke  column is:', avg_stroke)

#next, we replace missing values with calculated mean
df['stroke'].replace(np.nan, avg_stroke, inplace = True)
```

```python
#next, lets move on to the next missing value -'horsepower'

#first, we calculate and print the mean value to screen
avg_horsepower = df['horsepower'].astype('float').mean(axis =0)
print('Average of horsepower  column is:', avg_horsepower)

#next, we replace missing values with calculated mean
df['horsepower'].replace(np.nan, avg_horsepower, inplace = True)
```

```python
#lastly, lets move on to the next missing value -'peak-rpm'

#first, we calculate and print the mean value to screen
avg_peak_rpm = df['peak-rpm'].astype('float').mean(axis =0)
print('Average of peak_rpm  column is:', avg_peak_rpm)

#next, we replace missing values with calculated mean
df['peak-rpm'].replace(np.nan, avg_peak_rpm, inplace = True)
```

Now lets replace the missing values for the 'num-of-doors' column and also remove all the rows in the price column with missing price values. We are dropping missing price rows because, we are trying to predict the prices of cars, thus tempering with the values, or adjusiting price may later affect our model or any further analysis.

```python
#for num-of-doors, we will replace the 
#missing values with the value for the most category
```

To see which values are present in a particular column, we can use the ".value_counts()" method:

```python
#calling the values_counts() function on the 'num-of-doors' column
#to view values for various door categories

df['num-of-doors'].value_counts()
```

```python
#now lets use the '.idxmax()' method to 
#confirm the most common door type

df['num-of-doors'].value_counts().idxmax()
```

```python
# now let's replace the missing values with the most
#common car door number, whichis 'four'

df['num-of-doors'].replace(np.nan, 'four', inplace = True)
```

Finally, let's drop all rows that do not have price data:

```python
#lets drop the missing price values using the dropna() function

df.dropna(subset =['price'], axis =0, inplace = True)

#since we dropped the NaN price values, we need to reset
# the variable index

df.reset_index(drop =True, inplace =True)
```

```python
df.head()
```

```python
# lets take a look at the data types in dataset
# so we can convert them into appropriate formats

df.dtypes
```

```python
#now we can change the formats for the following variables
#
df[['bore', 'stroke', 'price', 'peak-rpm']] = df[[
    'bore', 'stroke', 'price', 'peak-rpm']].astype('float')
df[['normalized-losses']] = df[["normalized-losses"]].astype('int')
```

```python
df.dtypes
```

```python
# now we want to standardize the data
#we want to convert mpg to km

#to convert, L/100km = 235mpg, we divide 235 by the city-mpg values
df['city-L/100km'] = 235/df['city-mpg']

df.head()
```

```python
#to convert, L/100km = 235mpg, we divide 235 by 
#the highway-mpg values
df['highway-L/100km'] = 235/df['highway-mpg']

#now lets rename the columns
```

```python
df.head()
```

```python
df.rename(columns = {"'highway-L/100km'": 'highway-L/100km'}, inplace = True)
```

```python
df.head()
```

```python
#normalizing data values for width, length, height
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
```

```python
df[['length', 'width', 'height']].head()
```

```python
# creating bins to categorize horsepower values 
#into high, medium and low horsepower

#first
df['horsepower'] = df['horsepower'].astype('int', copy = True)
```

```python
df['horsepower'].dtypes
```

```python
# to create reflective bins, lets first view the
# distribution of horsepower values in the data
#using a histogram

from matplotlib import pyplot as plt
plt.hist(df['horsepower'])
plt.xlabel('Horsepower')
plt.ylabel('Number /Count')
plt.title('Horsepower Bins')
plt.show()
```

the above shows that majority of the cars sold in our data are between 50 to 150 horsepower.

```python
#first lets create a bin array
bins = np.linspace(min(df['horsepower']), 
                   max(df['horsepower']), 4)
bins
```

```python
# next lets determine the group names for the bins
groupnames = ['Low', 'Medium', 'High']
```

```python
#now lets apply pandas 'cut' function
#to determine which group each value of horsepower belong to

df['horsepower-binned'] = pd.cut(
    df['horsepower'], bins, labels = groupnames, 
    include_lowest = True)

#lets have a look at orginal and binned horsepower values
df[['horsepower', 'horsepower-binned']].head(20)
```

```python
#lets count how many highs, mediums and lows are there
df['horsepower-binned'].value_counts()
```

```python
# now lets plot the new horsepower values and see the distribution
plt.bar(groupnames, df['horsepower-binned'].value_counts())
plt.xlabel('Horsepower')
plt.ylabel('Number /Count')
plt.title('Horsepower Bins')
plt.show()
```

```python
#lets attempt to use a histogram to view 
#original horsepower values with a histogram, in 3 bins
a = (0, 1, 2)
plt.hist(df['horsepower'], bins = 3)
plt.xlabel('Horsepower')
plt.ylabel('Number /Count')
plt.title('Horsepower Bins')
plt.show()
```
