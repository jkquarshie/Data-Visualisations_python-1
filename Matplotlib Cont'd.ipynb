```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
plt.plot([2, 4, 6, 8], [4, 8, 12, 16], color ='red')

plt.xlabel('X', fontsize = 15, color = 'green')
plt.ylabel('Y = X-2', fontsize = 15, color = 'green')
```

```python
plt.plot(4, 8, 12, 16)
```

```python
x = np.linspace(start=0, stop= 10, num = 50)
y = np.sin(x)

plt.plot(x, y, color = 'red', label = 'Sine Curve')
plt.plot(x, np.cos(x), label = 'cosine curve')

plt.xlabel('X-value', fontsize = 15, color = 'green')
plt.ylabel('sin of X-value', fontsize = 15, color = 'green')
plt.title('This is Our New Curve - sin of x-value')
plt.legend()
```
