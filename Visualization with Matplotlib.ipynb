```python
import matplotlib.pyplot as plt #now import matplotlib as plt

plt.ion() #set the interactive mode or toggle it on
matplotlib.is_interactive() #check if interactive mode is now on; and yes, it is.
```

```python
plt.plot([2, 4, 6, 8], [4, 8, 12, 16]) #lets plot a simple chart to see the outcome
```

Now we can see our plot has several features, and buttons which we can use to zoom, shift or even save the plot
